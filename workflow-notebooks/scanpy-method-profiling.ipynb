```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.cm as cm
import matplotlib.colors as colors
from glob import glob
%matplotlib inline

import scanpy.api as sc
import math
import random
from scipy import stats
from timeit import timeit
from itertools import combinations
import phate

sns.set(style='whitegrid', palette='Set2', font_scale=1.3)
```

```python
# load data and cell type annotations
facs = glob('/Users/sbell/git/tabula-muris/00_data_ingest/00_facs_raw_data/FACS/*csv')
rawdata = pd.concat([pd.read_csv(f, header=0, index_col=0).T for f in facs[:15]])
```

```python
adata = sc.AnnData(rawdata.values, obs=rawdata.index.values, var=rawdata.columns.values)
adata.obs.index = rawdata.index.values
adata.obs_names = rawdata.index.values
adata.var.index = rawdata.columns.values
adata.var_names = rawdata.columns.values
cell_annotation = pd.read_csv('/Users/sbell/git/tabula-muris/00_data_ingest/03_tissue_annotation_csv/Heart_facs_annotation.csv', index_col=0) 
adata.obs['cell_type'] = cell_annotation.cell_ontology_class
```

```python
# do a cursory normalization of the data
sc.pp.filter_cells(adata, min_genes=100) # make the matrix less sparse
sc.pp.filter_genes(adata, min_cells=3)
sc.pp.normalize_per_cell(adata, counts_per_cell_after=1e4) # each cell sums to 1

filter_result = sc.pp.filter_genes_dispersion(adata.X, n_top_genes=10000, log=False)
adata._inplace_subset_var(filter_result.gene_subset)  # filter genes
sc.pp.log1p(adata) # log scale
sc.pp.scale(adata, max_value=10) # center + scale
```

```python
def time_per_dataset(fn, dataset_scales, title=None):
    time = { size: timeit(lambda : fn(dataset), number=1) for size, dataset in dataset_scales.items() }
    fig, ax = plt.subplots()
    x,y=list(zip(*time.items()))

    ax.plot(x, y, 'o-')
    ax.ticklabel_format(style='plain', axis='y')
    ax.set_xlabel('Number of cells')
    ax.set_ylabel('Execution time (seconds)')
    if title:
        ax.set_title(title)
    ax.set_xlim(0, 1.05*max(dataset_scales.keys()))
    plt.tight_layout()
    plt.show()
    
    return time
```

```python
## Run PCA and ask for the first 2 PCs
sc.tl.pca(adata, n_comps=2) 

## We'll run clustering once initially so we can use cluster-informed downsampling later
sc.pp.neighbors(adata,metric='euclidean', random_state=2018) 
sc.tl.louvain(adata, random_state=2018) # Then use the Louvain algorithm to identify clusters
adata.obs['louvain'] = adata.obs['louvain'].astype(str)

def downsample_by_cluster(adata, N, cluster_key = 'louvain'):
    '''Cluster-informed downsampling of cells in an anndata object
       adata: anndata object
       proportion: proportion of cells from each cluster to KEEP
       cluster_key: where to find cluster IDs in adata.obs[]
       returns adata[subsample,:]
    '''
    assert 1 < N < len(adata.obs_names)
    assert cluster_key in adata.obs.columns.values, 'Oops. Run clustering first with sc.pp.neighbors -> sc.tl.louvain'

    total_N = len(adata.obs_names)
    clusters = adata.obs.groupby('louvain')
    all_selected = []

    for cluster, cells in clusters.groups.items():
        cluster_size = len(cells)
        choose_k = int(N*(cluster_size/total_N))
        selected = random.sample(list(cells), k=choose_k)
        all_selected += selected

    return adata[all_selected, :]

### Downsample to 30%, 50%, 70%, and 100% of original data size.
total_N = len(adata.obs)
dataset_scales = { n : downsample_by_cluster(adata, n) for n in [5000,15000,45000] }
```

## Scanpy clustering & parameters  
### Scanpy includes only one method for clustering, `sc.tl.louvain`, which implements the louvain algorithm to assign cells to clusters based on the k-nearest-neighbors graph first built with `sc.pp.neighbors`.  
  
#### `sc.pp.neighbors` parameters to expose to user:  
* n_neighbors: int  
* method: str {'umap', 'gauss'}  

[?] metric: str {any metric accepted by scipy.spatial.distance.pdist}*

#### `sc.tl.louvain` parameters to expose to user:  
* resolution: float  

```python
pca_time = time_per_dataset(sc.pp.pca, dataset_scales, title='PCA')
neighbors_time = time_per_dataset(sc.pp.neighbors, dataset_scales, title='Neighbor graph')
louvain_time = time_per_dataset(sc.tl.louvain, dataset_scales, title='Cluster assignment')
```

## Trajectories  
### Scanpy supports 3 trajectory-related methods:  
  
### `sc.tl.dpt()`: pseudotime, cell ordering  
Params to expose:  
* `min_group_size`: float  
  
### `sc.tl.paga()`: minimum spanning tree amongst categories  
Params to expose:  
* `groups`: str (categorical metadata field)  
  
### `sc.tl.phate()`: layout based on trajectories-sensitive embedding  
Params to expose:  
* `k`: int  
* `a`: float  
* `gamma`: float  

```python
iroot = random.choice(range(min(dataset_scales.keys()))) ## set a root cell
for d in dataset_scales.values():
    d.uns['iroot'] = iroot

dpt_time = time_per_dataset(sc.tl.dpt, dataset_scales, title='diffusion pseudotime')
paga_time = time_per_dataset(sc.tl.paga, dataset_scales, title='paga (minimum spanning tree)')
```

```python
method_times = { 'pca': pca_time,
                 'neighbors': neighbors_time,
                 'louvain': louvain_time, 
                 'graph': graph_time,
                 'tsne': tsne_time,
                 'umap': umap_time, 
                 'diffmap': diffmap_time,
                 'dpt': dpt_time, 
                 'paga': paga_time}
method_times = pd.DataFrame(method_times, columns = ['neighbors', 'louvain', 'graph', 'tsne', 
                                                     'umap', 'pca', 'diffmap', 'dpt', 'paga']).T
sns.set(font_scale=1.2)
sns.heatmap(method_times, vmin=1, vmax=30, cmap=cm.BuGn, annot=True, linewidths=1)
```

```python
def cumulative_time(df, upstream, downstream):
    cumulative_df = df.copy().T
    upstream_vals = cumulative_df[upstream]
    for method in downstream:
        cumulative_df['%s+%s'%(upstream, method)] = cumulative_df[method] + upstream_vals
        del cumulative_df[method]
    return cumulative_df.T

time_with_pca = cumulative_time(method_times, 'pca', ['neighbors', 'tsne'])
time_with_neighbors = cumulative_time(time_with_pca, 'pca+neighbors', ['louvain', 'graph', 'umap', 'diffmap','paga'])
time_with_diffmap = cumulative_time(time_with_neighbors, 'pca+neighbors+diffmap', ['dpt'])
```

```python
fig, ax = plt.subplots(figsize=(6,6))
sns.heatmap(time_with_diffmap, vmin=1, vmax=30, cmap=cm.BuGn, annot=True, linewidths=1, ax=ax)
ax.set_title('Compute times upon regraph\nincluding all upstream dependencies')
```
